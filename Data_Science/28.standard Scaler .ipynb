```python
#Splitting dataset into input and output
x= Dataset1.drop("Outcome",axis=1)
x
```

```python
y=Dataset1.loc[:,['Outcome']]
y
```

```python
#Train_test_split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.75,random_state=42)
```

# Applying the standard scaler
# standard scaler only apply on input data (independent data)
# apply on x_train

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
#Learning on x_train
sc.fit(x_train)
```

```python
#Transform to SC -X_train,X_test
x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)
```
