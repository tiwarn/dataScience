```python
#how to find out number of male and female?
#palette is for colors (Reds,Purples,Blues,Greens and many more)
ax=sns.countplot (x='Gender',data=df,palette='Greys')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title('number of male and female')

```

```python
df['Gender'].value_counts()
```

```python
#how to find out number of orders
ax=sns.countplot(x='Orders',data=df,palette='Reds')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title('no of orders')
                 
```

```python
df['Orders'].value_counts()
```

```python
#how to find out the number of age group?
ax=sns.countplot(x='Age Group',data=df,palette='Blues')

for bar in ax.containers:
    ax.bar_label(bars)
plt.title('no of age group')
```

```python
df['Age Group'].value_counts()
```

```python
#Gender wise according to age group
ax=sns.countplot(x='Age Group',hue='Gender',data=df,palette='Reds')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Age group & Gender')
```

```python
#Number of marital status
ax=sns.countplot(x='Marital_Status',data=df,palette='Blues')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title('no of marital status')
```

```python
#occupations
ax=sns.countplot(x='Occupation',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Occupation')
```

```python
#RC is use for set_style or set_context
ax=sns.countplot(x='Occupation',data=df)

sns.set(rc={"figure.figsize":(5,25)})
    
for bar in ax.containers:
    ax.bar_label(bars)
    
plt.title('Occupation')
```

```python
df['Occupation'].value_counts()
```
