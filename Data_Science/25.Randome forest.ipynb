n_estimators:
This is the number of trees you want to build before taking the maximum voting or averages of predictions.
Higher number of trees give you better performance but makes your code slower.

Entropy:
Entropy is nothing but the measures of disorder or uncertainity in a given dataset.it is commonly used to evaluate the quality of a model and its ability to make accurate predictions. 

Calculating entropy 2 class problem
When data has only 2 class labels (yes/no)
entropy y(s) = -p(yes)  log2  p(yes) â€“ p(no) log2 p(no)


Calculating entropy 3 class problem
When data has 3 class labels (yes/no/maybe)
entropy y(s) = -p(yes)  log2  p(yes) â€“ p(no) log2 p(no)- p(maybe) log2 p(maybe)

1.for a 2 class problem the min entropy is 0 and maximun entropy is 1.
2.for more then  2 class problem the min entropy is 0 and maximun entropy is grater than 1.

```python
# training random forest classification on training set
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators = 3, criterion = "entropy")
```

```python
model.fit(x_train,y_train)
```
