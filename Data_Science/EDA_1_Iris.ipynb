***bivariate analysis (only take 2 features - "SepalLengthCm","SepalWidthCm")**

```python
# facetgrid - a method in which we can plot data in the form of a grid like structure, use for visualization.
# hue = hue parameter determines which column in the data frame should be used for colour encoding.
sns.FacetGrid(df,hue="Species",height=5).map(plt.scatter,"SepalLengthCm","SepalWidthCm")
```

```python
#Add legend
sns.FacetGrid(df,hue="Species",height=5).map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend();
```

***multivariate analysis ( use all features)**


```python
# pairplot is use for multivariate analysis.
sns.pairplot(df,hue="Species",height=4)
```

```python
sns.pairplot(df,x_vars=["SepalLengthCm","PetalWidthCm"],y_vars=['Species'],height=4,aspect=1);
```
