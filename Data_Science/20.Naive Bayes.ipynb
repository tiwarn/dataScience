```python
# seprate independent variable and dependent variable
inputs=df.drop("Survived",axis="columns")
inputs.head()
```

```python
target=df.Survived
target.head()
```

```python
# convert sex column in numerical with help of dummie
dummies=pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
#Add dummies in to dataframe
inputs=pd.concat([inputs,dummies],axis=1)
inputs.head()
```

```python
#Find null value in any columns
inputs.columns[inputs.isna().any()]
```

```python
#to find null value
inputs.Age[:20]
```

```python
## fill null value help of the mean
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.head(20)
```

```python
#Split data in to train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.3)
```

```python
#Create model
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
#Train model
model.fit(x_train,y_train)
```
