#model building 


#In adaboost you can change the individual model apart from decision tree


```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# inside adaboost algorithm, running logistic regression.
# HERE WE CREATE 150 LOGISTIC REGREESION COMBINATION AND make ADABOOSTCLASSIFIER.
model2=AdaBoostClassifier(estimator=LogisticRegression(),n_estimators=150)
```

```python
model2.fit(x_train,y_train)
```

```python
y_pred=model2.predict(x_test)
y_pred
```

```python
#Accuracy_score
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred)*100)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
