```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.datasets import load_iris
from tensorflow import keras
import pandas as pd
```

```python
#loading the dataset
iris=load_iris()
iris
```

```python
#Input
x=pd.DataFrame(iris.data,columns=iris.feature_names)
x
```

```python
#Output
y=iris.target
y
```

#Now we are going to apply one hot encoding on our output values, so that our model would give equal importance to all the classes

```python
encoder = OneHotEncoder(sparse=False)
y = encoder.fit_transform(y.reshape(-1, 1))
```

```python
y
```

```python
#Train Test Split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.83,random_state=42)
```

#Now we are going to create the Neural Network

```python
# Initialise the model
model1 = Sequential()

# Add input layer(4 features) and a hidden layer with 10 neuron
model1.add(Dense(10, input_shape = (4,), activation = 'relu'))

# Add output layer( 3 classes - S, V, V)
model1.add(Dense(3, activation = 'softmax'))

# compile the model
model1.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

# fit the model
model1.fit(x_train, y_train, epochs = 100)
```

```python
#Normal approach
```

```python
model2 = Sequential([
    Dense(10, input_shape = (4,), activation = 'relu'),
    Dense(3, activation = 'softmax')
])

model2.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

model2.fit(x_train, y_train, epochs = 100)
```

```python
#3rd Approch
```

```python
model3 = keras.Sequential([
    keras.layers.Dense(10, input_shape = (4,), activation = 'relu'),
    keras.layers.Dense(3, activation = 'softmax')
])

model3.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

model3.fit(x_train, y_train, epochs = 100)
```
