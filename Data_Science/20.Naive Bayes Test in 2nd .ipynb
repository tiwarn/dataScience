```python
inputs=df.drop(["Survived","Sex"],axis="columns")
inputs.head()
               
```

```python
Target=df.Survived
Target.head()
```

```python
y=df.Sex
y.head()
```

```python
y=y.map({"male":0,"female":1})
y.head()
```

```python
inputs=pd.concat([inputs,y],axis=1)
inputs
```

```python
# find null value in any columns
inputs.columns[inputs.isna().any()]
```

```python
# to find null value
inputs.Age[:20]
```

```python
# fill null value help of the mean

inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.head(10)
```

```python
# split data into train_test_split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs,Target, test_size=0.3)
```

```python
# create model
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
# train model
model.fit(x_train,y_train)
```

```python
# to compare y_test to predict x_test
y_test[:10]
```

```python
model.predict(x_test[:10])
```

```python
model.score(x_test,y_test)*100
```
