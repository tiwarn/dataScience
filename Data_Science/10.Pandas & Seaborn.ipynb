Python Pandas Data Structure
The Pandas provides two data structures for processing the data, i.e., Series and DataFrame, which are discussed below:
1) Series
It is defined as a one-dimensional array that is capable of storing various data types.
We can easily convert the list, tuple, and dictionary into series , using "series' method.
A Series cannot contain multiple columns. It has one parameter:

```python
#Series ( before creating a series, Firstly we have to import the numpy module and then use array() function in the program.
import pandas as pd
import numpy as np
info = np.array(['p','a','n','d','a','s'])
a = pd.Series(info)
print(a)
```

2) DataFrame
It is a widely used data structure of pandas and works with a two-dimensional array with labelled axes (rows and columns).
DataFrame is defined as a standard way to store data and has two different indexes, i.e., row index and column index.

Series object attributes

Series.index	Defines the index of the Series.
Series.shape	It returns a tuple of shape of the data.
Series.dtype	It returns the data type of the data.
Series.size	    It returns the size of the data.
Series.empty	It returns True if Series object is empty, otherwise returns false.
Series.hasnans	It returns True if there are any NaN values, otherwise returns false.
Series.nbytes	It returns the number of bytes in the data.
Series.ndim	    It returns the number of dimensions in the data.

```python
# Retrieving Index array and data array of a series object
import pandas as pd
x = pd.Series(data=[0,1,2,3,4,5,6,7])
y = pd.Series(data=[11.2,18.6,22.5],index=['a','b','c'])

print(x.index)
print(x.values)
print(y.index)
print(y.values)
              
```

```python
# Retrieving Types (dtype) and Size of Type (itemsize)
import pandas as pd
a=pd.Series(data=[1,2,3,4])
b=pd.Series(data=[4.9,8.2,5.6])
print(a.dtype)
print(a.size)
print(b.dtype)
print(b.size)
```

```python
#Retrieving Shape
import pandas as pd
a=pd.Series(data=[1,2,3,4])
b=pd.Series(data=[4.9,8.2,5.6])
print(a.shape)
print(b.shape)
```

```python
#Checking Emptiness and presence of  NoNs
import pandas as pd
a=pd.Series(data=[1,2,3,np.NaN])
b=pd.Series(data=[4.9,8.2,5.6])
c=pd.Series()

print(a.empty,b.empty,c.empty)
print(a.hasnans,b.hasnans,c.hasnans)
print(len(a),len(b))
print(a.count(),b.count())
```
