```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#here we are getting error because 403 that mean server denies your request 
requests.get('https://www.ambitionbox.com/list-of-companies?page=1')
```

```python
#here you can see the error in text
requests.get('https://www.ambitionbox.com/list-of-companies?page=1').text
```

```python
#Now we will pass the header that mean API will understand, this request coming from brouser
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
#Now we will pass the header that mean API will understand, this request coming from brouser
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'} # Updated User-Agent
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers,timeout=10,verify=False).text
```
